```python
import pandas as pd
```

```python
data = pd.read_csv("training.csv")
```

```python
data
```

```python
x = data.drop("target",axis=1)
```

```python
x
```

```python
y = data["target"]
```

```python
y
```

# Decison Tree

```python
from sklearn.metrics import mean_squared_error
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model1 = DecisionTreeClassifier()
```

```python
model1 = model1.fit(x, y)
```

```python
test_data = pd.read_csv("test.csv")
```

```python
test_data
```

```python
x_test = test_data.drop("target",axis =1)
```

```python
y_test = test_data["target"]
```

```python
model1_predict = model1.predict(x_test)
```

```python
model1_accuracy = mean_squared_error(y_test, model1_predict)
```

```python
model1_accuracy
```

# Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model2 = RandomForestClassifier()
```

```python
model2 = model2.fit(x,y)
```

```python
model2_predict = model2.predict(x_test)
```

```python
model2_accuracy = mean_squared_error(y_test, model2_predict)
```

```python
model2_accuracy
```

# Native Bayes

```python
from sklearn.naive_bayes import GaussianNB
```

```python
model3 = GaussianNB()
```

```python
model3 = model3.fit(x,y)
```

```python
model3_predict = model3.predict(x_test)
```

```python
model3_accuracy = mean_squared_error(y_test, model3_predict)
```

# Support Vector Machine

```python
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
```

```python
from sklearn.svm import LinearSVC
```

```python
model4 = make_pipeline(StandardScaler(), LinearSVC(random_state=0, tol=1e-5))
```

```python
model4 = model4.fit(x,y)
```

```python
model4_predict = model4.predict(x_test)
```

```python
model4_accuracy = mean_squared_error(y_test, model4_predict)
```

```python
model4_accuracy
```
